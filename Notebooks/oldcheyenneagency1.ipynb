```python
# List of file paths
filepaths = [
    "USGS_LPC_SD_FY17_NRCS_Lidar_QSI_2017_D17_14TLQ396985.laz",
    "USGS_LPC_SD_FY17_NRCS_Lidar_QSI_2017_D17_14TLQ397986.laz",
    "USGS_LPC_SD_FY17_NRCS_Lidar_QSI_2017_D17_14TLQ397985.laz",
    "USGS_LPC_SD_FY17_NRCS_Lidar_QSI_2017_D17_14TLQ396986.laz"
]

# Lists to store coordinate arrays
x_all, y_all, z_all = [], [], []

# Use a loop to read and store all coordinates
for path in filepaths:
    las = laspy.read(path)
    x_all.append(las.x)
    y_all.append(las.y)
    z_all.append(las.z)

# Concatenate into full arrays
x = np.concatenate(x_all)
y = np.concatenate(y_all)
z = np.concatenate(z_all)

print(f"Loaded {len(x)} total points from {len(filepaths)} tiles.")
```

```python
print("X range:", x.min(), "-", x.max())
print("Y range:", y.min(), "-", y.max())

# Check coordinates 
if not (x.min() > 0 and y.min() > 0):
    print("Warning: Some coordinates have non-positive values, check for errors in data.")
    
# Try in case the arrays are empty 
try:
    if len(x) == 0 or len(y) == 0:
        raise ValueError("Empty x or y array detected — check input files or preprocessing.")
except ValueError as e:
    print("Error:", e)

# check that x and y arrays are the same length
if len(x) != len(y):
    print("Warning: x and y arrays are of different lengths.")
```

```python
# Combine coordinates and classifications
x = np.concatenate([las.x])
y = np.concatenate([las.y])
z = np.concatenate([las.z])
classification = np.concatenate([las.classification])
```

```python
classes, counts = np.unique(classification, return_counts=True)
print("Classification codes and counts:")
for cls, cnt in zip(classes, counts):
    print(f"Class {cls}: {cnt} points")
```

```python
# Create water mask
water_mask = classification == 9

# Set up side-by-side plots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 7), sharex=True, sharey=True)

# Full LiDAR plot
sc1 = ax1.scatter(x, y, c=z, cmap="terrain", s=1)
ax1.set_title("All LiDAR Points")
ax1.set_xlabel("X")
ax1.set_ylabel("Y")
plt.colorbar(sc1, ax=ax1, label="Elevation (Z)")

# Water-only plot
sc2 = ax2.scatter(x[water_mask], y[water_mask], c=z[water_mask], cmap="Blues", s=1)
ax2.set_title("Water-Classified Points (Class 9)")
ax2.set_xlabel("X")
plt.colorbar(sc2, ax=ax2, label="Elevation (Z)")

plt.suptitle("LiDAR Point Cloud vs Water-Classified Points", fontsize=16)
plt.tight_layout()
plt.show()
```

```python
# Use the combined classification array
water_mask = classification == 9

# Get water points from combined arrays
x_water = x[water_mask]
y_water = y[water_mask]
z_water = z[water_mask]

# Plot elevation histogram of water-classified points
plt.figure(figsize=(10, 6))
plt.hist(z_water, bins=200, color='skyblue', edgecolor='black')
plt.title("Elevation Histogram of Water-Classified Points")
plt.xlabel("Elevation (Z)")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

```python
from sklearn.cluster import DBSCAN

# Water points
sample_size = 10000
idx = np.random.choice(len(x_water), size=sample_size, replace=False)
coords = np.column_stack((x_water[idx], y_water[idx], z_water[idx]))

# Apply DBSCAN
clustering = DBSCAN(eps=2.0, min_samples=30).fit(coords)
labels = clustering.labels_

# Plot clusters
plt.figure(figsize=(10, 8))
plt.scatter(coords[:, 0], coords[:, 1], c=np.where(labels == -1, -1, labels), cmap='tab20', s=1)
plt.title("Clusters in Water — Possible Submerged Structures")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()

# Print Z-range for each cluster 
unique_clusters = set(labels)
for cluster_id in unique_clusters:
    if cluster_id == -1:
        continue  # skip noise
    cluster_z = coords[labels == cluster_id][:, 2]
    print(f"Cluster {cluster_id}: Z range = {cluster_z.min():.2f} to {cluster_z.max():.2f}, size = {len(cluster_z)}")
```

```python
import laspy
import pandas as pd
import geopandas as gpd
from shapely.geometry import Point

from pathlib import Path
data_dir = Path("USGS_LPC_SD_FY17_NRCS_Lidar_QSI_2017_D17_14TLQ396986.laz")  # or "." if files are in current directory
cheyenne_agency1 = data_dir / "USGS_LPC_SD_FY17_NRCS_Lidar_QSI_2017_D17_14TLQ396985.laz"

las = laspy.read(data_dir)
x, y, z = las.x, las.y, las.z
```

```python
# Ensure x, y, z are NumPy arrays and of equal length
import numpy as np
from shapely.geometry import Point
import geopandas as gpd

# Just to be safe, cast to arrays (if not already)
x = np.array(x)
y = np.array(y)
z = np.array(z)

# Check that all arrays are the same length
try:
    assert len(x) == len(y) == len(z)
    
    # Create list of shapely Points
    geometry = [Point(xy) for xy in zip(x, y)]
    
    # Create GeoDataFrame
    gdf = gpd.GeoDataFrame({'elevation': z}, geometry=geometry)
    
    # Set CRS (South Dakota is usually EPSG:26914 - UTM Zone 14N)
    gdf.set_crs("EPSG:26914", inplace=True)
    
    print("GeoDataFrame successfully created.")
except AssertionError:
    print("Error: x, y, and z arrays must be the same length.")
except Exception as e:
    print("Unexpected error during GeoDataFrame creation:", e)
```

```python
from scipy.interpolate import griddata
import numpy as np
import numpy.ma as ma

# Limit number of points for interpolation
max_points = 10000
if len(x_water) > max_points:
    idx = np.random.choice(len(x_water), size=max_points, replace=False)
    x_sample = x_water[idx]
    y_sample = y_water[idx]
    z_sample = z_water[idx]
else:
    x_sample = x_water
    y_sample = y_water
    z_sample = z_water

# Interpolation grid
grid_x, grid_y = np.mgrid[
    min(x_sample):max(x_sample):500j,
    min(y_sample):max(y_sample):500j
]

# Interpolate Z values
grid_z = griddata(
    (x_sample, y_sample),
    z_sample,
    (grid_x, grid_y),
    method='linear'
)

# Mask invalids
masked_grid_z = ma.masked_invalid(grid_z)

# Plot the DEM
plt.figure(figsize=(10, 8))
plt.imshow(
    masked_grid_z.T,
    extent=(min(x_sample), max(x_sample), min(y_sample), max(y_sample)),
    origin='lower',
    cmap='terrain'
)
plt.colorbar(label="Elevation (Z)")
plt.title(" Water-Classified Points")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
